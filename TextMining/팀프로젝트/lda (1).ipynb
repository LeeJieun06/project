```python
import numpy as np
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import gensim
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis

# Load the data
data = pd.read_excel('/content/drive/MyDrive/[텍스트분석] 팀 프로젝트/dataset/article_df_final.xlsx', engine='openpyxl')

# Preprocessing
data['main'] = data['main'].str.replace(pat=r'[^가-힣ㄱ-하-ㅣ\\s]', repl=r' ', regex=True)
data['main'] = data['main'].str.replace(r'[0-9]+', repl=r' ', regex=True)

# Tokenize the text
tokenized_text = data['main'].apply(lambda x: re.findall(r'\w{2,}', x.lower()))

# Create dictionary and corpus
dictionary = gensim.corpora.Dictionary(tokenized_text)
corpus = [dictionary.doc2bow(text) for text in tokenized_text]

# Build LDA model
lda_model = gensim.models.LdaModel(corpus=corpus,
                                   id2word=dictionary,
                                   num_topics=10,
                                   random_state=1,
                                   passes=10)

# Visualize the topics
vis = gensimvis.prepare(lda_model, corpus, dictionary)
pyLDAvis.display(vis)
```
