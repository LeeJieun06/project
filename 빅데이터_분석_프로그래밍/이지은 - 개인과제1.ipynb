# p.194

```python
#194쪽
#1
mpg=pd.read_csv('./Data/mpg.csv')
mpg.loc[[9,13,57,92],'drv']='k'
mpg.loc[[28,42,128,202],'cty']=[3,4,39,42]

mpg['drv'].isin(['k']).value_counts()
k=mpg['drv'].isin(['k'])
mpg['drv']=np.where(k,np.nan,mpg['drv'])

mpg['drv'].isin(['k']).value_counts()#결측치 확인
```

```python
#2
iqr= mpg['cty'].quantile(.75)-mpg['cty'].quantile(.25)
maxrange=mpg['cty'].quantile(.75)+iqr*1.5
minrange=mpg['cty'].quantile(.25)-iqr*1.5
print(maxrange,"   ", minrange)

import seaborn as sns
sns.boxplot(data=mpg,y='cty')#이상치 제거전

mpg['cty']=np.where((mpg['cty']> maxrange) | (mpg['cty']< minrange), np.nan ,mpg['cty'])
sns.boxplot(data=mpg,y='cty')#이상치 제거후 박스플롯
```

```python
#3
mpg.dropna(subset=['drv','cty']).groupby('drv').agg(avg_cty=('cty','mean'))
```

# p.211

```python
#211쪽
#1
mpg=pd.read_csv('./Data/mpg.csv')

car=mpg.query('category=="suv"')
car=car.groupby('manufacturer')\
.agg(avg=('cty','mean'))\
.sort_values('avg',ascending=False)\
.head()


car.reset_index(inplace=True)#interpret오류때문에 인덱스 리셋시킴


sns.barplot(data=car,x='manufacturer',y='avg');
```
